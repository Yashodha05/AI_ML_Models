```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np
df=pd.read_csv('C:/Users/MANASA N S/OneDrive/ドキュメント/MINE PERSONAL/DATASETS CSV FORMATS/auto-mpg/auto-mpg.csv') 
print(df.head(10),"\n","\n")
print("Min : ",df['mpg'].min(),"\n","\n")
print("Max : ",df['mpg'].max(),"\n","\n")
print("Quantile",df['mpg'].quantile([.1, .25, .5, .75]),"\n","\n") 
```

```python
df.isnull().sum()
```

```python
# code to get all cars with 8 cylinder ---
eight_cylinder_car = df[df['cylinders'] == 8]
print(eight_cylinder_car)
```

```python
# cars manufactured in each year ----
cars_by_yr = df.groupby('model year')['car name'].count()
cars_by_yr
```

```python
# display details of SUV cars. (A car with HP > 160 and weight > 3500)
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')

suv_cars = df[(df['horsepower']>160)  & (df['weight'] > 3500)]
print(suv_cars)
```

```python
# Display details of Muscle cars . (A car with HP > 160, displacement > 320 and weight > 3500 )
muscle_cars = df[(df['horsepower']>160) & (df['displacement'] > 320) & (df['weight'] > 3500)]
print(muscle_cars)
```

```python
print(plt.hist(x=df['mpg'])) 
```

```python
print(plt.scatter(x='weight',y='mpg',data=df))
```

```python
print(df['origin'].value_counts().plot(kind='bar')) 
```

```python
print(sns.boxplot(df['mpg']))
```
